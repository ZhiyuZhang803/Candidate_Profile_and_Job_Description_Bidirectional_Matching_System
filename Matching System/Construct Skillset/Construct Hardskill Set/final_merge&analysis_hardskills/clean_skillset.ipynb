```python
import pandas as pd
```

```python
df = pd.read_excel("final_skill_table.xlsx")
df.head()
```

```python
# delete the repetitive items and empty items
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df_drop = df[df['skill_set']!="['']"]
df_drop = df_drop.reindex()
df_drop.head()
```

```python
# transfer skillset column to list
skill_set_list = [i.strip("[").rstrip("]").split(",") for i in df_drop['skill_set']]
skill_set_list = [(h.strip("[").strip("'").strip("'").strip(" '") for h in i) for i in skill_set_list]
# drop rep items
skill_set = [list(set(i)) for i in skill_set_list]
# replace the original column
df_drop['skill_set'] = skill_set
```

```python
# clean the data: delete the \ in the dataset
temp_skill = []
for i in df_drop["skill_set"]:
    single_skill = []
    for h in i:
        if '\\' in h.replace('"',''):
            single_skill.append(h.replace('"','').replace("\\",""))
        elif h =="":
            continue
        else:
            single_skill.append(h)
    temp_skill.append(single_skill)

temp_skill[:5]
```

```python
df_drop['skill_set'] = temp_skill
df_drop.head()
```

```python
# expand skill list 0: split the words with / (Unix/Linux -> Unix, Linux)
temp_skill = []
for i in df_drop["skill_set"]:
    single_skill = []
    for h in i:
        if '/' in h:
            single_skill.append(h)
            for m in h.split("/"):
                single_skill.append(m)
        else:
            single_skill.append(h)
    temp_skill.append(single_skill)

temp_skill[:5]
```

```python
df_drop['skill_set'] = temp_skill
df_drop.head()
```

```python
# expand skill list 3: split the words if abbreviation is attached at the tail of the word (Visual Basic Studio VBA -> Visual Basic Studio, VBA)
def more_than_two_upper(word):
    count = 0
    for i in word:
        if i == i.upper():
            count = count+1
        else:
            continue
    if count >= 2:
        return True
    else:
        return False
```

```python
temp_skill = []
for i in df_drop["skill_set"]:
    single_skill = []
    for h in i:
        for single_word in h.split(" "):
            if more_than_two_upper(single_word) == True:
                single_skill.append(h.replace("single_word","").strip().replace("  "," "))
                single_skill.append(single_word)
        single_skill.append(h)
    temp_skill.append(single_skill)

temp_skill[1]
```

```python
df_drop['skill_set'] = temp_skill
df_drop.head()
```
